```python
# Weight init
import numpy as np
import matplotlib.pyplot as plt

n_feat = 784
n_hidden = 256
np.random.seed(190131)
a = np.random.randn(1, n_feat)
b = np.random.randn(n_feat, n_hidden)
out = np.dot(a, b).reshape(n_hidden)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].hist(out)
ax[0].set_title("Distribution of inputs to activation function")
ax[1].hist(np.tanh(out))
ax[1].set_title("Distribution of outputs of tanh function")
```

```python
# Normalized
out = out - np.mean(out)
out = out / np.std(out)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].hist(out)
ax[0].set_title("Distribution of inputs to activation function")
ax[1].hist(np.tanh(out))
ax[1].set_title("Distribution of outputs of tanh function")
```

```python
# Normal distribution
out = np.random.normal(loc=0, scale=1, size=n_hidden)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].hist(out)
ax[0].set_title("Distribution of inputs to activation function")
ax[1].hist(np.tanh(out))
ax[1].set_title("Distribution of outputs of tanh function")
```

```python
# Glorot
scale = 2/(n_feat + n_hidden)
out = np.random.normal(loc=0, scale=scale, size=n_hidden)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].hist(out)
ax[0].set_title("Distribution of inputs to activation function")
ax[1].hist(np.tanh(out))
ax[1].set_title("Distribution of outputs of tanh function")
```
